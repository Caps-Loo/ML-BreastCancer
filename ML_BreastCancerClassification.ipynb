# **1. Import Library**

# **2. Read Dataset**

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Memuat dataset Breast Cancer
cancer = load_breast_cancer()

# Menyusun fitur dan target
X = cancer.data
y = cancer.target

# Mengonversi ke DataFrame
df_X = pd.DataFrame(X, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')

# Menampilkan DataFrame fitur dan target
print("Fitur DataFrame:")
print(df_X.head())  # Menampilkan 5 baris pertama fitur
print("\nTarget DataFrame:")
print(df_y.head())  # Menampilkan 5 baris pertama target
```

```python
df_X
```

```python
df_y
```

```python
# Gabungkan fitur dan target dalam satu DataFrame
df = pd.concat([df_X, df_y], axis=1)

df.head(10)
```

```python
df.info()
```

```python
df['target'].unique()
```

```python
df.describe()
```

# **3. Exploratory Data Analysis (EDA)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan distribusi kelas target (benign dan malignant)
plt.figure(figsize=(6, 4))
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Distribusi Kelas Target (Benign vs Malignant)')
plt.xlabel('Target')
plt.ylabel('Jumlah')
plt.tight_layout()  # Menghindari label yang terpotong
plt.show()

# Plot distribusi fitur seperti 'mean radius' menggunakan histogram
plt.figure(figsize=(6, 4))
sns.histplot(df['mean radius'], kde=True, color='skyblue')
plt.title('Distribusi Fitur Mean Radius')
plt.xlabel('Mean Radius')
plt.ylabel('Frekuensi')
plt.tight_layout()  # Menghindari label yang terpotong
plt.show()
```

# **4. Split Data**

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
```

# Decision Tree

# **Train the Model**

```python
from sklearn.tree import DecisionTreeClassifier

# Membuat dan melatih model Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

# **Predict & Evaluate**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Memprediksi dan mengevaluasi
y_pred = model.predict(X_test)

accuracy_dt = accuracy_score(y_test, y_pred)

print("Laporan Klasifikasi:")
print(f"Akurasi: {accuracy_dt * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix")
plt.show()
```

# **Analysis Results**


```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test, y_pred, target_names=cancer.target_names))
```
